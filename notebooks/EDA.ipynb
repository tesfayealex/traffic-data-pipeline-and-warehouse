```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("../data/trial_dataset.csv",sep=";",index_col=False)
```

```python
original_df = df
df = original_df.iloc[: , :10]
df
```

```python

df.info()
```

```python
sns.displot(data=df, x=" traveled_d", color="blue",
                    kde=True, height=7, aspect=2)
plt.title(f'Distribution of traveled distance', size=20, fontweight='bold')
plt.show()
```

```python
sns.displot(data=df, x=" avg_speed", color="red",
                    kde=True, height=7, aspect=2)
plt.title(f'Distribution of speed', size=20, fontweight='bold')
plt.show()
```

```python
plt.figure(figsize=(12, 7))
sns.countplot(data=df, x=" type")
plt.title(f'Distribution of vehicle type', size=20, fontweight='bold')
plt.show()
```

```python
test_df = df[[" type" , " traveled_d" , " avg_speed" , " speed" , " time"]]
# test_df.rename(columns={"Motercycle": 1 , "Car": 2 , "Bus":3 , "Taxi":4 , "Medium_Vehicle": 5 , "Heavy Vehicle": 6}, inplace = True) 
test_df.loc[test_df[' type'] == " Motorcycle", ' type'] = 1
test_df.loc[test_df[' type'] == " Car", ' type'] = 2
test_df.loc[test_df[' type'] == " Bus", ' type'] = 3
test_df.loc[test_df[' type'] == " Taxi", ' type'] = 4
test_df.loc[test_df[' type'] == " Medium Vehicle", ' type'] = 5
test_df.loc[test_df[' type'] == " Heavy Vehicle", ' type'] = 6
test_df[" type"] = test_df[" type"].astype("float64")
test_df
```

```python
plt.figure(figsize=(12, 7))
sns.heatmap(test_df, annot=True, cmap='viridis', vmin=0,
                    vmax=1, fmt='.2f', linewidths=.7, cbar=False)
plt.title("Correlation matrix of Important columns", size=18, fontweight='bold')
plt.show()
```
